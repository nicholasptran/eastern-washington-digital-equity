```r
library(psych)
library(tidyverse)

data <- read.csv("app/data/z_score_data.csv")
data <- data[3:27]
```

```r
dist_data <- dist(data, method = 'euclidean')

hclust_avg <- hclust(dist_data, method = 'average')

cut_avg <- cutree(hclust_avg, k = 5)
```

```r
plot(hclust_avg)
rect.hclust(hclust_avg, k = 5, border = 1:30)
abline(h = 5, col = 'red')
head(data)
```
