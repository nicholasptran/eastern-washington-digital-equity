```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

data1 = pd.read_csv("../app/data/combined_data.csv")
data = data1.drop(columns=["NAME", "GEOID", "median_income"])
```

```python
data.scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
weights = [0.45368933, 0.35268022, -0.50360634, -0.04384384,
           -0.26747852, 0.85924065, 1.06137118, -0.90171551]
```

```python
data.scaled["index"] = data.scaled @ weights
```

```python
data.scaled
```

```python
final_df = data1[["NAME", "GEOID"]].join(data.scaled)
```

```python
final_df.to_csv("../app/data/temp.csv", index=False)
```
