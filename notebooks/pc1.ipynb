```r
library(tidyverse)

data <- read.csv("../data/standardized_data.csv") %>% select(-GEOID, -tract)
```

```r
library(tidyverse)
library(gridExtra)
head(data, 10)
```

```r
data.cov <- cov(data)
data.cov
data.eigen <- eigen(data.cov)
str(data.eigen)
data.eigen$vectors[,1:27])
```

```r
phi <- data.eigen$vectors
phi <- -phi
```

```r
phi
```

```r
PC1 <- as.matrix(data) %*% phi[,1]
```

```r
PVE <- data.eigen$values / sum(data.eigen$values)
round(PVE, 2)
```

```r
PVEplot <- qplot(c(1:27), PVE) +
    geom_line() +
    xlab("Principal Component") +
    ylab("PVE") +
    ggtitle("Scree Plot") +
    ylim(0,1)
```
